#dataset

```
from fastbook import *
from fastai.vision.widgets import *
```

```
path = Path("/content/drive/My Drive/Airplane_dataset")
```

```
plane = DataBlock(blocks = (ImageBlock, CategoryBlock), #image...和cate...都是內建函數 代表資料刑事
                  get_items = get_image_files, #get_image_file 取得資料方式
                  get_y = parent_label, #parent_label找網上一層路徑名稱
                  splitter = RandomSplitter(valid_pct = 0.1, seed = 7414), #data讀進來後怎麼分成 train 和 validation，random隨機分類 ，seed 是代表出來的結果都依樣
                  item_tfms = Resize(448), #先裁減大小再縮小
                  batch_tfms = aug_transforms(mult = 1.5) # 影像增強
                  )
```

```
dls = plane.dataloaders(path, bs = 8) #類似圖片轉矩陣 這裡是轉成tensor 
```

```
dls.show_batch()
```

```
dls.valid.show_batch()
```

#train

```
learner = cnn_learner(dls, resnet34, metrics = error_rate) #用cnn dls 放dataloader ,resnet152 選架構 , metrics 是最後一步判斷model的好壞及準確度 ,loss function 一半是在判斷train時model的好壞 
```

```
learner.summary()
```

```
recorder = learner.lr_find() #找平緩的步伐
```

```
torch.cuda.empty_cache()
```

```
learner.fit_one_cycle(12, lr_max = 1e-3) #目的是找最低點兒不適區域最低點，4 代表4次跌代，lr_max找最適合的learning rate
```

```
recorder = learner.lr_find() #找平緩的步伐
```

```
learner.save("/content/drive/My Drive/Airplane_dataset/stage1")
```

```
learner.load("/content/drive/My Drive/Airplane_dataset/stage1")
```

```
learner.unfreeze()
```

```
interp = ClassificationInterpretation.from_learner(learner) #valid data拿出來看model表現
```

```
interp.plot_confusion_matrix() #把confusion matrix 畫出來
```

```
interp.plot_top_losses(30, nrows = 30)
```
